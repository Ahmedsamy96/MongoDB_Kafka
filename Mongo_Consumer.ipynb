```python
from kafka import KafkaConsumer

print("Kafka Consumer Application Started ... ")

KAFKA_BOOTSTRAP_SERVERS_CONS = 'localhost:9092'
kafka_producer_obj = KafkaConsumer("Kafka-To-Talend",bootstrap_servers=KAFKA_BOOTSTRAP_SERVERS_CONS,auto_offset_reset="latest",enable_auto_commit=True,group_id="Talend_Group")

for message in kafka_producer_obj:
    print("Message : ", message)
```
